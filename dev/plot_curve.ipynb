```python
import plotext as pt
import numpy as np


x = np.linspace(0, 10, 1000)
y = np.sin(x)

pt.plot(x, y)
pt.show()
```

```python

def plot(x, y):
    pt.clt()
    pt.cld()
    pt.plot(x, y)
```

```python
import time
x = np.linspace(0, 1, 10)
y = np.sin(x)

pt.clc()

for point in range(len(x)):
    plot(x[:point], y[:point])
    time.sleep(1)
    pt.show()
pt.clear_plot()
```

```python
def edit_previous_line(text: str, line: int = 1, *, return_str: bool = False):
    # out = f'\033[{line}F\033[K\033[a{text}'
    out = f'\033[{line}F\033[K{text}'
    for i in range(line - 1):
        out += '\n'
    if return_str:
        return out
    else:
        print(out)
```
